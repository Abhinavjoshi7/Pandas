```python
import matplotlib.pyplot as plt
```

## If only one parameter is passed to the plot function, the parameter is considered the y value and it is plotted using x, starting from 0, going to n
    [2,4,6,8] -> [0,1,2,3] 0 for 2, 1 for 4 and so on.

```python
plt.plot([2,4,6,8])
```

```python
plt.plot([2,4,2,1,6,8])
```

## plot takes x as the first axis and y as the second axis

```python
salaries = [55000, 60000, 70000, 85000, 100000]
ages = [20,25,30,40,45]
plt.plot(ages, salaries)
```

## plt.plot(x,y) will not show a visualization in a normal python file, we would need to use the .show() function to see the results.

```python
import numpy as np
nums = np.arange(5)
```

```python
plt.plot(nums, nums)
plt.plot(nums, nums*2)
plt.plot(nums, nums*3)
```

# Matplotlib terminology
## Figure is the higest level visual object, it is like a canvas. A figure can have mulitple visuals. 
## Each visual in a figure is called axes, it has x axis, y axis, data etc. Axes can be grouped in a figure.

```python
plt.figure()
```

## plt.figure() creates a new figure and activates it, whatever is added after the plt.figure() it will be added to the newly create figure

```python
plt.figure()
plt.plot(nums, nums)
plt.plot(nums, nums*2)
```

```python
plt.figure()
plt.plot(nums, nums)
plt.figure()
plt.plot(nums, nums*2)
```

```python
plt.figure(figsize=(2,6))
plt.plot(nums, nums)
plt.figure(figsize=(8,3))
plt.plot(nums, nums)
```

```python
plt.figure(figsize=(2,6), dpi=20)
plt.plot(nums, nums)
```
