```python
houses = pd.read_csv('data/kc_house_data.csv')
```

```python
houses
```

# We can also use a boolean expresion with loc

```python
many_bedrooms = houses['bedrooms'] >= 10
```

```python
houses.loc[many_bedrooms, ['bedrooms']] = 'Too Many!'
houses[many_bedrooms]
```

```python
high_quality_houses = houses['grade'] > 12
good_view_houses = houses['view'] == 4
houses[good_view_houses & high_quality_houses]
```

```python
houses['luxury_level'] = 'normal'
```

```python
houses
```

```python
houses.loc[good_view_houses & high_quality_houses, ['luxury_level']] = 'Luxury'
```

```python
houses[good_view_houses & high_quality_houses]
```
