## Bar Plots

```python
plants = ['Spinach', 'Turnip', 'Rhubarb', 'Broccoli', 'Kale']
died = [10, 25, 5, 30, 21]
germinated = [74, 88, 56, 69, 59]
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, germinated, width=0.5)
plt.title('Germination Rates')
plt.ylabel('Number of Seedings')
plt.xlabel('Plant Names')
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, width=0.5, color='red')
plt.title('Germination Rates')
plt.ylabel('Number of Dead Seedings')
plt.xlabel('Plant Names')
```

```python
plt.figure(figsize=(6,6))
plt.bar(plants, died, width=0.3, color='red')
plt.bar(plants, germinated, width=0.2, color='green',align='edge')
```

## Stacked bar chart

```python
plt.figure(figsize=(6,8))
plt.bar(plants, died, color='red', bottom=germinated, label='Failed Seedlings')
plt.bar(plants, germinated, color='green', label='Germinated Seedlings')
plt.legend(shadow=True, frameon=True, facecolor='white')
plt.title('Greenhouse Seedlings')
```

```python
plt.figure(figsize=(6,6))
plt.barh(plants, died, color='red')
```

```python
first_class = titanic[titanic['pclass'] == 1]['age']
second_class = titanic[titanic['pclass'] == 2]['age']
third_class = titanic[titanic['pclass'] == 3]['age']
# Alpha is used to set the transparency
plt.hist(first_class, label='1st Class', alpha=0.5)
plt.hist(second_class, label='2st Class', alpha=0.5)
plt.hist(third_class, label='3st Class', alpha=0.5)
plt.legend()
```

## Working with subplots

```python
nums = np.arange(5)
plt.figure(figsize=(10,4))
# Sets the title of the the entire figure
plt.suptitle('Our First Subplot', fontsize=30)

# Here 1,3,1 means 1 row, 3 columns, and the 1/index means first visual
plt.subplot(1,3,1)
plt.title('X')
plt.plot(nums, nums)

plt.subplot(1,3,2)
plt.title('X Squared')
plt.plot(nums, nums**2, color='red')

plt.subplot(1,3,3)
plt.title('X Cubed')
plt.plot(nums, nums**3, color='green')

#Automatically adjusts the padding and spacing
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(4,10))
# Sets the title of the the entire figure
plt.suptitle('Our First Subplot', fontsize=30)

# Here 1,3,1 means 1 row, 3 columns, and the 1/index means first visual
plt.subplot(3,1,1)
plt.title('X')
plt.plot(nums, nums)

plt.subplot(3,1,2)
plt.title('X Squared')
plt.plot(nums, nums**2, color='red')

plt.subplot(3,1,3)
plt.title('X Cubed')
plt.plot(nums, nums**3, color='green')

#Automatically adjusts the padding and spacing
plt.tight_layout()
plt.show()
```
