```python
# Calculate how many days the dataset spans
lowest_date  = kc_houses.sort_values('date')['date'].head(1)

```

```python
highest_date  = kc_houses.sort_values('date')['date'].tail(1)
highest_date
```

```python
dataset_span = highest_date - lowest_date
dataset_span
```
