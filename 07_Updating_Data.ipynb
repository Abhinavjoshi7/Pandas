```python
import pandas as pd
```

```python
titanic = pd.read_csv('data/titanic.csv')
titanic
```

```python
titanic.sex.replace('female', 'F')
```

```python
titanic['sex'].replace(['female', 'male'], ['F', 'M'], inplace=True)
titanic
```

```python
# When you want to replace something with None, use the list syntax even if you are only replacing one.
```

```python
titanic['age'].replace(['?'], [None], inplace=True)
```

```python
titanic
```
